```python
import pickle

# Load the DataFrame from the pickle file
with open('compiled_data.pkl', 'rb') as file:
    compiled_data = pickle.load(file)
```

```python
normalized_data=compiled_data.iloc[0]['normalized_data']
beta_waves=compiled_data.iloc[0]['beta']
epochs=compiled_data.iloc[0]['epochs']
time=compiled_data.iloc[0]['raw_time']
channels=compiled_data.iloc[0]['channel_id']
```

```python
import numpy as np
white_trials=[]
black_trials=[]
for eventsi in epochs:
    print(eventsi)
    print(eventsi[0][1])
    print(eventsi[0][0])

    if eventsi[0][1]==119:
        gate_open=np.where(time>eventsi[0][0])
        white_trials.append(time[gate_open[0][0]])
    elif eventsi[0][1]==98:
        gate_open=np.where(time>eventsi[0][0])
        black_trials.append(time[gate_open[0][0]])
```

```python
#Plotting the data

import matplotlib.pyplot as plt

fig, ax =plt.subplots(figsize=(20, 5))
ax.plot(time, beta_waves)
# Plot vertical lines for each value in white_trials and black_trials
for i, (w, b) in enumerate(zip(white_trials, black_trials)):
    if i == 0:
        ax.axvline(w, color='grey', linestyle='--', label='White Trial')
        ax.axvline(b, color='black', linestyle='--', label='Black Trial')
    else:
        ax.axvline(w, color='grey', linestyle='--')
        ax.axvline(b, color='black', linestyle='--')
ax.set_title('Normalized data {}'.format(channels))
ax.set_xlabel('Time')
ax.set_ylabel('uV')
ax.legend()

plt.show()
```
