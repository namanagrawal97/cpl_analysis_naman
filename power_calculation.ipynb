```python
import numpy as np
def calculate_power(signal_array):
    """
    Calculate the power of each signal in the given array.
    
    Parameters:
    signal_array (numpy.ndarray): 2D array where each row is a signal.
    
    Returns:
    numpy.ndarray: 1D array containing the power of each signal.
    """
    # Initialize an empty list to store power values
    power_values = []
    
    # Iterate over each row in the array
    for signal in signal_array:
        # Calculate the power of the signal
        power = np.sum(signal ** 2) / len(signal)
        # Append the power value to the list
        power_values.append(power)
    
    # Convert the list to a NumPy array and return
    return np.array(power_values)
```
