```python
import pandas as pd

filter_data=pd.read_excel(r"C:\Users\sinha\Dropbox\CPLab\filecheck.xlsx")
filter_data = filter_data.dropna(axis=1, how='all')
filter_data = filter_data.dropna(axis=0, how='all')

filter_data = filter_data.drop(filter_data.columns[-1],axis=1)

filter_data.columns=['file','LFP1_AON', 'LFP2_AON', 'LFP3_AON', 'LFP4_AON','LFP1_vHp','LFP2_vHp','date' ]
```

```python
import os
filter_data_files=list(filter_data['file'])
filter_base_names_without_ext = [os.path.splitext(os.path.basename(file))[0] for file in filter_data_files]
print(filter_base_names_without_ext)
filter_data['file']=filter_base_names_without_ext
```

```python
import glob
files = glob.glob(r'C:\Users\sinha\Dropbox\CPLab\all_data_mat_filtered\*.mat')
print(files)
```

```python
print(len(files))
print(filter_data.shape)
```

```python
channel_names=['LFP1_AON', 'LFP2_AON', 'LFP3_AON', 'LFP4_AON','LFP1_vHp','LFP2_vHp']
import h5py
import os
not_here_list=[]
for file in files:
    f=h5py.File(file, 'r+')
    try:
        base_name = os.path.basename(file)
        base_name_without_ext = os.path.splitext(base_name)[0]
        #print(base_name_without_ext)
        if base_name_without_ext in filter_base_names_without_ext:
            print(base_name_without_ext,'yes')
            row = filter_data.loc[filter_data['file'] == base_name_without_ext]
            #print(row)
            for column in row.columns:
                value = row[column].values[0]
                if value == 'no':
                    print(f"Column '{column}' has value 'no'")
                    if column in f.keys():
                        del f[column]
                        print(f"Deleted key '{column}' from file '{file}'")
                elif value == 'ok':
                    print(f"Column '{column}' has value 'ok'")
        
        else:
            print(base_name_without_ext, 'this file not here')
            not_here_list.append(base_name_without_ext)
    finally:
        f.close()
print(not_here_list)
```
